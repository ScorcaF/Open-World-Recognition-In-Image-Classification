```python
from datetime import datetime
import sys
import numpy as np
import pickle


np.set_printoptions(threshold=sys.maxsize)

def save_logs(logs,RANDOM_SEED, NUM_EPOCHS, RUN_NAME, BATCH_SIZE):

  date = datetime.now().strftime("%Y_%m_%d-%I:%M_%p")
  file_name = ''+ RUN_NAME +'_'+ date +'.txt'

  # print run info + logs
  file = open('/content/drive/MyDrive/MLDL Group project/File colab/File progetto/logs/'+file_name, "w")
  file.write(" ## "+ 'number of iterations: ' + str(len(RANDOM_SEED)) + " ## " +
             "seeds: "+str(RANDOM_SEED) + " ## " + "\n" +" #### "+
             "type of run: "+ RUN_NAME  + " #### "+ "\n" + " ## "+
             "number of epochs: "+str(NUM_EPOCHS)+" ## " +
             "batch size: "+str(BATCH_SIZE)+ " ## " +'\n')
  file.write( f'logs : '+ '\n' + str(logs)+ '\n')
  file.close()

  #save the logs object:
  obj_name = ''+ RUN_NAME +'_'+ date +'.pickle'
  file_to_store = open('/content/drive/MyDrive/MLDL Group project/File colab/File progetto/logs/'+obj_name, "wb")
  pickle.dump(logs, file_to_store)
  file_to_store.close()

  


```
